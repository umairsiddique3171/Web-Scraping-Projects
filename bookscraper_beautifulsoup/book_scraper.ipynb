```python
# importing essential libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
url = "https://books.toscrape.com/"  # website url
```

```python
page = requests.get(url)  # requesting page content
```

```python
soup = BeautifulSoup(page.text,'html')  # saving page content in html format
```

```python
soup.find_all('h3') # finding all data with h3 tag
```

```python
# extracting title and img link
for card in soup.find_all('h3'):
    print(card.text)
    print(card.a.get('href'))
    print('')
```

```python
# extracting price
for card in soup.find_all('p',class_ = 'price_color'):
    print(card.string[1:])
```

```python
# extracting text
for card in soup.find_all('p',class_ = 'instock availability'):
    print(card.text.strip())
```

```python
# extracting rating
for card in soup.find_all('p',class_ = 'star-rating'):
    print(card['class'][1])
```

```python
# combining them all
for x,y,z,t in zip(soup.find_all('h3'),soup.find_all('p',class_ = 'price_color'),soup.find_all('p',class_ = 'instock availability'),soup.find_all('p',class_ = 'star-rating')):
    print(x.text)
    print('https://books.toscrape.com/' + x.a.get('href'))
    print(y.string[1:])
    print(z.text.strip())
    print(t['class'][1])
    print('')
```

```python
# saving data in form of list
b = []
for x,y,z,t in zip(soup.find_all('h3'),soup.find_all('p',class_ = 'price_color'),soup.find_all('p',class_ = 'instock availability'),soup.find_all('p',class_ = 'star-rating')):
    a = []
    a.append(x.text)
    a.append('https://books.toscrape.com/'+ x.a.get('href'))
    a.append(y.string[1:])
    a.append(z.text.strip())
    a.append(t['class'][1])
    b.append(a)
print(b)
```

```python
data = np.array(b) # converting data into array
```

```python
data.shape
```

```python
# creating dataframe from array
df = pd.DataFrame(data,columns = ['Book Name','Img Link','Price','Availability','Rating (out of five)'] ) 
```
